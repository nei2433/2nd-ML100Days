1.1. 讀取 txt 檔, 請讀取 text file 

```python
import pandas as pd

#讀檔
data = []
with open("./data/example.txt", 'r') as f:
    for line in f:
        line = line.replace('\t', ',') # 將每句Tab鍵以逗號斷句
        line = line.replace('\n', '').split(',') # 將每句最後的 /n 取代成空值後，再以逗號斷句
        data.append(line)
data
```

1.2 將所提供的 txt 轉成 pandas dataframe

```python
df = pd.DataFrame(data) #轉換為dataframe
df
```

```python
df.drop(range(2,5),axis=1,inplace=True) #拉掉None的欄
df.head(5)
```

2. 從所提供的 txt 中的連結讀取圖片，請讀取上面 data frame 中的前 5 張圖片

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
top5=df[1].head(5).tolist() #先將前五個連結存為list
top5
```

```python
import skimage.io as skio #顯示圖片(註: 第五個連結找不到圖片)
for i in top5:
    img1 = skio.imread(i)
    plt.imshow(img1)
    plt.show()
```
