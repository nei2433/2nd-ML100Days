```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
app_train = pd.read_csv('./data/application_train.csv')
print(app_train.head())
print(app_train.shape)
```

```python
# 將只有兩種值的類別型欄位, 做 Label Encoder, 計算相關係數時讓這些欄位可以被包含在內
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

# 檢查每一個 column
for col in app_train:
    if app_train[col].dtype == 'object':
        # 如果只有兩種值的類別型欄位
        if len(list(app_train[col].unique())) <= 2:
            # 就做 Label Encoder, 以加入相關係數檢查
            app_train[col] = le.fit_transform(app_train[col])            
print(app_train.shape)
print(app_train.head())
```

```python
# 受雇日數為異常值的資料, 另外設一個欄位記錄, 並將異常的日數轉成空值 (np.nan)
app_train['DAYS_EMPLOYED_ANOM'] = app_train["DAYS_EMPLOYED"] == 365243
app_train['DAYS_EMPLOYED'].replace({365243: np.nan}, inplace = True)

# 出生日數 (DAYS_BIRTH) 取絕對值 
app_train['DAYS_BIRTH'] = abs(app_train['DAYS_BIRTH'])
```

```python
# 觀察相關係數
CORR=app_train.corr()['TARGET']
```

```python
print(CORR.sort_values().head(15)) #最小前15項
CORR_HEAD15=list(CORR.sort_values().head(15).index.values)
```

```python
print(CORR.sort_values(ascending=False).head(16)) #最大前15項
CORR_TAIL15=list(CORR.sort_values(ascending=False).head(16).index.values)
CORR_TAIL15.remove('TARGET')#必須移除TARGET
```

```python
#繪製最小前15項之箱型圖(因TARGET為二元變數)
for i in CORR_HEAD15:
    app_train.boxplot(column=i,by='TARGET')
plt.show()
```

```python
#繪製最大前15項之箱型圖(因TARGET為二元變數)
for i in CORR_TAIL15:
    app_train.boxplot(column=i,by='TARGET')
plt.show()
```
